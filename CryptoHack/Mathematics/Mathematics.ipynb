### Chinese Remainder Theorem

````python
from sympy import mod_inverse

congruences = [
    (2, 5),   # x ≡ 2 mod 5
    (3, 11),  # x ≡ 3 mod 11
    (5, 17)   # x ≡ 5 mod 17
]

N = 1
for _, n in congruences:
    N *= n
x = 0
for a_i, n_i in congruences:
    N_i = N // n_i
    y_i = mod_inverse(N_i, n_i)
    x += a_i * N_i * y_i

x %= N
print(f"The solution is: x ≡ {x} mod {N}")
````

````python
def brute_force(sequence, primes):
    for p in primes:
        for x in range(2, p):
            if all((sequence[i] * x) % p == sequence[i+1] for i in range(len(sequence) - 1)):
                return p, x
    return None, None

sequence = [588, 665, 216, 113, 642, 4, 836, 114, 851, 492, 819, 237]
primes = [i for i in range(851, 999) if sympy.isprime(i)]
prime, base = brute_force(sequence, primes)
print(f"crypto{{{prime},{base}}}")
````
